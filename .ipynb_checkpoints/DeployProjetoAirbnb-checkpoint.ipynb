```python
import pandas as pd
import streamlit as st # streamlit é a biblioteca Python que torna a criação de aplicativos da web interativos e visualizações de dados em Python mais fácil e rápida
import joblib # biblioteca python que serve para salvar objetos python em arquivos, tambem serve para ler arquivos gerados pela mesma biblioteca

# usaremos a biblioteca streamlit para podermos criar uma especie de 'mini site', 
# para que o usuario final possa preencher as informacoes para que o nosso projeto de data science possa prever o preço do imóvel.
# usaremos a biblioteca joblit para facilitar a leitura do nosso projeto de data science.

# pegou todas as colunas do dataframe do projeto Airbnb e separou por tipos de dados: numericos(x_numericos), 
#verdadeiro e falso(x_tf) e categorias(x_listas)  Lembrando que a coluna price nao esta nesses dicionarios , pois ela é a resposta final do nosso projeto.

x_numericos = {'latitude': 0, 'longitude': 0, 'accommodates': 0, 'bathrooms': 0, 'bedrooms': 0, 'beds': 0, 'extra_people': 0,
               'minimum_nights': 0, 'ano': 0, 'mes': 0, 'n_amenities': 0, 'host_listings_count': 0}

x_tf = {'host_is_superhost': 0, 'instant_bookable': 0}

x_listas = {'property_type': ['Apartment', 'Bed and breakfast', 'Condominium', 'Guest suite', 'Guesthouse', 'Hostel', 'House', 'Loft', 'Outros', 'Serviced apartment'],
            'room_type': ['Entire home/apt', 'Hotel room', 'Private room', 'Shared room'],
            'cancellation_policy': ['flexible', 'moderate', 'strict', 'strict_14_with_grace_period']
            }


dicionario = {} # criou um dicionario vazio para editar as variaveis dummies de x_listas
for item in x_listas:
    for valor in x_listas[item]:
        dicionario[f'{item}_{valor}'] = 0 # usou o for para concatenar o nome de cada item de x_listas com o valor de cada item




for item in x_numericos: # usou for para percorrer as informacoes do tipo numericas (x_numericos)
    if item == 'latitude' or item == 'longitude': 
        valor = st.number_input(f'{item}', step=0.00001, value=0.0, format="%.5f") # usou st.number_input para inserir os botoes de preenchimento em cada valor de x_numericos, e o nome desses botoes será exatamente o nome das informacoes de x_numericos, step: é o quanto o botao irá andar, value: valor inicial, format: é o tanto de casas decimais que o valor terá (streamlit)
    elif item == 'extra people':
        valor = st.number_input(f'{item}', step=0.01, value=0.0) 
    else:
        valor = st.number_input(f'{item}', step=1, value=0)
    x_numericos[item] = valor
    
    
for item in x_tf:   # usou for para percorrer as informacoes do tipo verdadeiro ou falso (x_tf)
    valor = st.selectbox(f'{item}', ('Sim','Não')) #usou st.selectbox para inserir os botoes de seleção em cada valor de x_tf, as opcoes que irao aparecer sao 'Sim' e 'Não' e o nome desses botoes será exatamente o nome das informacoes de x_tf (streamlit) 
    if valor == 'Sim':
        x_tf[item] = 1 # sim = 1
    else:
        x_tf[item] = 0 # nao = 0
        
        
for item in x_listas:  # usou for para percorrer as informacoes do tipo categorias (x_listas)
    valor = st.selectbox(f'{item}', x_listas[item]) #usou st.selectbox para inserir os botoes de seleção em cada valor de x_listas, as opcoes que irao aparecer sao todas opcoes em colchetes da x_listas e o nome desses botoes será exatamente o nome das informacoes de x_listas (streamlit) 
    dicionario[f'{item}_{valor}'] = 1
    
botao = st.button('Prever valor do imóvel') # este botao será responsavel por prever o preço do imovel (streamlit)

# precisamos juntar todas os dicionarios que contem os nomes das colunas todo sem 1 só (comando update)
if botao: # se botao for acionado
    dicionario.update(x_numericos) # python
    dicionario.update(x_tf)    # python
  
 # foi criado um dataframe contendo todos os nomes desse dicionario nas colunas:
    valores_x = pd.DataFrame(dicionario, index=[0]) # pandas

# as colunas do dataframe 'valores_x' devem ser exatamente iguais a do nosso modelo mas sem a coluna preço
    dados = pd.read_csv('dados.csv') # leu a base de dados tratada
    colunas = list(dados.columns)[1:-1] # colocou o nome das colunas e uma lista e retirou as colunas de 'price' e 'unnamed'
    valores_x = valores_x[colunas]

# carregando o nosso modelo que foi criado com a biblioteca joblib:
    modelo = joblib.load('modelo.joblib')
    preco = modelo.predict(valores_x) # resposta do nosso projeto , ou seja , a previsao do preco
    st.write(preco[0]) # st.write servirá para escrever a resposta do preço na tela


```
